```python
%%sql
--datatypes enz later in view
CREATE TABLE asn_transactie(
	boekings_datum nvarchar(50) NOT NULL,
	rekening_nr_self nvarchar(50) NOT NULL,
	rekening_nr nvarchar(50) NULL,
	naam_rekening nvarchar(50) NULL,
	adres nvarchar(1) NULL,
	postcode nvarchar(1) NULL,
	plaats nvarchar(1) NULL,
	valuta_rekening nvarchar(50) NOT NULL,
	saldo_voor_mutatie float NOT NULL,
	valuta_mutatie nvarchar(50) NOT NULL,
	bedrag float NOT NULL,
	journaal_datum nvarchar(50) NOT NULL,
	valuta_datum nvarchar(50) NOT NULL,
	transactie_code smallint NOT NULL,
	transactie_afk nvarchar(50) NOT NULL,
	transactie_volgnr int NOT NULL,
	betalingskenmerk nvarchar(50) NULL,
	omschrijving nvarchar(200) NULL,
	afschriftnr tinyint NOT NULL

)
```

```python

%%sql
DROP VIEW vw_asn_transactie;
Create VIEW vw_asn_transactie AS
SELECT
       case when omschrijving like '%Kosten gebruik%' AND naam_rekening IS NULL THEN 'bankkosten'
            when omschrijving like '%CCV*Laadpaal%'                             THEN 'autokosten'
            when omschrijving like '%HYUNDAI%'                                  THEN 'autokosten'
            when omschrijving like '%refund%'                                   THEN 'autokosten'
            when naam_rekening like '%SHELL%'                                   THEN 'autokosten'
            when naam_rekening like '%RANDSTAD%'                                THEN 'inkomsten'
            when naam_rekening like '%Amazon%'                                  THEN 'kantoorkosten'
            when naam_rekening like '%AMZN%'                                    THEN 'kantoorkosten'
            when naam_rekening like '%bol.com%'                                 THEN 'kantoorkosten'
            when naam_rekening like '%Coolblue%'                                THEN 'kantoorkosten'


            when naam_rekening like 'Framework Computer Inc via Framework Computer LLC' THEN 'kantoorkosten'
            when naam_rekening like 'FUTURUMSHOP.NL B.V.'                               THEN 'autokosten' --drager
            when naam_rekening like 'Gemeente Nijmegen via ING'                         THEN 'autokosten' --parkeren
            when naam_rekening like 'Heuts.nl via MultiSafepay'                         THEN 'autokosten'
            when naam_rekening like 'Kamer van Koophandel via CCV Group BV'             THEN 'overig'
            when naam_rekening like '%Belsimpel%'                                       THEN 'telefoonkosten'
            when naam_rekening like '%CHING%'                                           THEN 'prive'
            when naam_rekening = 'Belastingdienst'                                      THEN 'belasting'
            when naam_rekening = 'CENTRAAL BEHEER'                                      THEN 'verzekeringen'
            --N.V. Schadeverz.-Mij. Bo --geen btw

            else 'overig/prive/uitzoeken'


        end as categorie
      ,case when bedrag < 0 then 'uitgaven'
        else 'inkomsten' end as transactie_type
      ,cast(boekings_datum as date) as boekings_datum
      ,omschrijving
      ,bedrag
      ,bedrag/1.21  as bedrag_ex_btw
      ,bedrag/1.21*.21 as bedrag_btw
      /*
      ,format(bedrag, 'N', 'nl-NL') as bedrag
      ,format(bedrag/1.21 , 'N', 'nl-NL') as bedrag_ex_btw
      ,format(bedrag/1.21*.21, 'N', 'nl-NL') as bedrag_btw
      ,bedrag as bedrag_bron
      */
      ,naam_rekening
      ,rekening_nr
      ,date_part('quarter',cast(boekings_datum as date)) as kwartaal
      ,betalingskenmerk
      --,saldo_voor_mutatie
      --,journaal_datum
      --,valuta_datum
      --,transactie_code
      --,transactie_afk
      --,transactie_volgnr

      --,afschriftnr
      --,rekening_nr_self
  FROM asn_transactie
  where 1=1
    and bedrag <> 0
 -- order by naam_rekening
 ;
```

```python
%%sql


    --inkomsten
SELECT categorie, sum(bedrag) as bedrag_totaal, sum(bedrag_ex_btw) as omzet, sum(bedrag_btw) as omzetbelasting
  FROM vw_asn_transactie
  where 1=1
    and kwartaal = 4
    and categorie NOT IN ('belasting','bankkosten', 'prive','overig','overig/prive/uitzoeken')
    and categorie = 'inkomsten'
  group by categorie
  order by categorie;


```

```python
%%sql
--voorbelasting
-- shell heeft geen standaard 21% btw, buitenland, handmatig van factuur
drop view if exists vw_asn_transactie_2024_q1;
create view vw_asn_transactie_2024_q1 as
select
    *,
    CASE WHEN omschrijving  = '24F064084' THEN 0.84
         WHEN omschrijving = 'APS00000000003622126' THEN 16.74 + 7.01
         WHEN omschrijving = '24F134766' THEN 0.84
         WHEN omschrijving = '24F241170' THEN 4.15
         WHEN omschrijving = 'payatt3OtboRKfnk3HlqaY06P2KMSZ 0220887389455038 Framework Computer Inc' THEN 2263-1870.27
         else abs(bedrag_btw)
        end as voorbelasting_gecorrigeerd
        ,
from vw_asn_transactie
where 1=1
    and transactie_type = 'uitgaven'
    and categorie not in ('belasting','bankkosten')
    and year(boekings_datum) = 2024
order by 1

```
